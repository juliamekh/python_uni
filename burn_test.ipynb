```python
import csv

with open('questions.csv', 'r') as file:
    reader = csv.reader(file)
    questions = list(reader)


answers_child = {}
answers_adult = {}
answers_parent = {}


for i, question in enumerate(questions):
    print(f"Вопрос {i+1}: {question[0]}")
    answer = int(input("Введите ответ (1-10): "))
    
    
    if i+1 in [1, 4, 7, 10, 13, 16, 19]:
        answers_child[question[0]] = answer
    elif i+1 in [2, 5, 8, 11, 14, 17, 20]:
        answers_adult[question[0]] = answer
    elif i+1 in [3, 6, 9, 12, 15, 18, 21]:
        answers_parent[question[0]] = answer

sum_child = sum(answers_child.values())
sum_adult = sum(answers_adult.values())
sum_parent = sum(answers_parent.values())

print(' ')
if sum_child > sum_adult and sum_child > sum_parent:
    print("Категория: Ребенок.")
elif sum_adult > sum_child and sum_adult > sum_parent:
    print("Категория: Взрослый.")
else:
    print("Категория: Родитель.")
  
print('Тест не является полностью достоверным. Для полноценного анализа личности необходимо обратиться к психологу.')
```
